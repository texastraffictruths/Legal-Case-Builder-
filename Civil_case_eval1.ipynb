```python
import re
import json

# Load or create case data
try:
    with open(f"{case_name}.json") as f:
        case_data = json.load(f)
except FileNotFoundError: # Catch the specific error if the file doesn't exist
    case_data = {"summary": "", "violations": "", "remedies": "", "evidence": "", "qa": []}

# Extract questions from the AI's output (simple split)
# Ensure ai_questions variable exists before trying to use it
if 'ai_questions' in globals() and ai_questions:
    questions = re.findall(r"\d+[\).\-] (.+)", ai_questions)
else:
    # Handle the case where ai_questions might not be defined or is empty
    print("No AI questions were generated in the previous step.")
    questions = [] # Initialize questions as an empty list to avoid errors later

# Loop through questions and ask user
for q in questions:
    print(f"\nü§ñ {q}")
    user_answer = input("üßë Your answer: ")

    # Have AI rephrase answer for clarity
    # Ensure the 'client' object is available in this cell's scope
    if 'client' in globals():
        clarify_response = client.chat.completions.create(
            model="meta-llama/llama-3.2-11b-vision-instruct:free",
            messages=[
                {"role": "system", "content": "Rephrase the user's legal answer into clear, formal legal language."},
                {"role": "user", "content": f"Question: {q}\nAnswer: {user_answer}"}
            ]
        )
        cleaned = clarify_response.choices[0].message.content
    else:
        print("Error: OpenAI client is not initialized.")
        cleaned = user_answer # Use the user's raw answer if client is not available

    # Try to classify where the answer belongs
    field = "qa"
    q_lower = q.lower()
    if "violation" in q_lower or "right" in q_lower:
        field = "violations"
    elif "evidence" in q_lower or "proof" in q_lower:
        field = "evidence"
    elif "remedy" in q_lower or "outcome" in q_lower:
        field = "remedies"

    # Append or update case_data based on the determined field
    if field in ["violations", "remedies", "evidence"]:
        # For single-value fields, replace the content
        case_data[field] = cleaned
    else:
        # For 'qa', append to the list
        case_data["qa"].append({"question": q, "answer": cleaned})


    print(f"‚úÖ Saved to: {field}")
    print(cleaned)

    # Save case file again
    with open(f"{case_name}.json", "w") as f:
        json.dump(case_data, f, indent=4) # Added indent for readability

print("\nüíæ Case answers stored and organized.")
```

```python
from docx import Document
from docx.shared import Pt, Inches
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.oxml.ns import qn
from docx.oxml import OxmlElement
from datetime import datetime
import os
import json  # Import json here

def create_formatted_docx(title_key, content_key, doc_type="general", suffix="docx"): # Changed suffix default to docx
    # Load case
    # The try/except block for loading the case was missing here, adding it for robustness
    try:
        with open(f"{case_name}.json") as f:
            case = json.load(f)
    except FileNotFoundError:
        print(f"Error: Case file '{case_name}.json' not found. Please create or load a case first.")
        return # Exit the function if the case file isn't found
    except NameError:
        print("Error: 'case_name' is not defined. Please run the cell that defines the case name.")
        return # Exit the function if case_name isn't defined


    metadata = case.get("metadata", {})
    content = case.get(content_key, None)

    if not content:
        print(f"‚ùå No content found under key: {content_key}")
        return

    # Count version
    existing = [d for d in case.get("documents", []) if d["type"] == doc_type]
    version = len(existing) + 1
    filename = f"{doc_type}_v{version}.{suffix}"

    # Start DOCX document
    doc = Document()

    # Set margins
    sections = doc.sections
    for section in sections:
        section.top_margin = Inches(1)
        section.bottom_margin = Inches(1)
        section.left_margin = Inches(1)
        section.right_margin = Inches(1)

    # Add court header
    caption = metadata.get("court_division", "IN THE UNITED STATES DISTRICT COURT\nFOR THE SOUTHERN DISTRICT OF TEXAS\nGALVESTON DIVISION")
    style = doc.styles['Normal']
    font = style.font
    font.name = 'Times New Roman'
    font.size = Pt(12)

    para = doc.add_paragraph()
    para.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    for line in caption.split("\n"):
        para.add_run(line + "\n").bold = True

    # Add case style and docket info
    case_style = metadata.get("case_style", "[Your Name] v. [Defendants]")
    docket = metadata.get("docket_number", "Civil Action No. ___________")

    para = doc.add_paragraph()
    para.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    para.add_run("\n" + case_style + "\n").bold = True
    para.add_run(docket + "\n").italic = True

    doc.add_paragraph("\n")  # Space

    # Add content with spacing
    for line in content.strip().split("\n"):
        if line.strip():
            p = doc.add_paragraph(line.strip())
            p.paragraph_format.space_after = Pt(12)

    # Add footer with page number
    section = doc.sections[0]
    footer = section.footer
    footer_para = footer.paragraphs[0]
    footer_para.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
    run = footer_para.add_run()
    # Correct way to add page number field in python-docx
    fldChar1 = OxmlElement('w:fldChar')
    fldChar1.set(qn('w:fldCharType'), 'begin')
    instrText = OxmlElement('w:instrText')
    instrText.set(qn('xml:space'), 'preserve')
    instrText.text = "PAGE \\* MERGEFORMAT" # Use MERGEFORMAT switch
    fldChar2 = OxmlElement('w:fldChar')
    fldChar2.set(qn('w:fldCharType'), 'end')
    separator = OxmlElement('w:r')
    t = OxmlElement('w:t')
    t.text = " of "
    separator.append(t)
    page_num = OxmlElement('w:fldChar')
    page_num.set(qn('w:fldCharType'), 'begin')
    page_instr = OxmlElement('w:instrText')
    page_instr.set(qn('xml:space'), 'preserve')
    page_instr.text = "NUMPAGES \\* MERGEFORMAT" # Use NUMPAGES field
    page_num_end = OxmlElement('w:fldChar')
    page_num_end.set(qn('w:fldCharType'), 'end')

    run._r.append(fldChar1)
    run._r.append(instrText)
    run._r.append(fldChar2)
    run._r.append(separator) # Add " of " separator
    run._r.append(page_num)
    run._r.append(page_instr)
    run._r.append(page_num_end)


    # Save DOCX
    doc.save(filename)
    print(f"‚úÖ Formatted document saved as: {filename}")

    # Track in case file
    # Ensure case data is not empty before attempting to add documents
    if case:
        if "documents" not in case:
            case["documents"] = []

        case["documents"].append({
            "title": f"{title_key} v{version}",
            "filename": filename,
            "type": doc_type,
            "version": version,
            "timestamp": datetime.utcnow().isoformat()
        })

        # Ensure json is imported before dumping
        import json
        with open(f"{case_name}.json", "w") as f:
            json.dump(case, f, indent=4) # Added indent for readability

        from google.colab import files
        files.download(filename)

# EXAMPLE USAGE:
# create_formatted_docx("Federal Complaint", "federal_complaint_draft", doc_type="complaint")
# create_formatted_docx("Affidavit of Fact", "affidavit_of_fact", doc_type="affidavit")
```

```python
from docx import Document
from docx.oxml import OxmlElement
from docx.oxml.ns import qn
import os

def add_toc_to_docx(docx_path):
    if not os.path.exists(docx_path):
        print("‚ùå File not found.")
        return # This return statement and the following lines were indented incorrectly

    doc = Document(docx_path)
    toc_para = doc.paragraphs[0].insert_paragraph_before()

    # Add TOC field
    run = toc_para.add_run()
    fldChar1 = OxmlElement('w:fldChar')
    fldChar1.set(qn('w:fldCharType'), 'begin')

    instrText = OxmlElement('w:instrText')
    instrText.set(qn('xml:space'), 'preserve')
    instrText.text = 'TOC \\o "1-3" \\h \\z \\u'

    fldChar2 = OxmlElement('w:fldChar')
    fldChar2.set(qn('w:fldCharType'), 'separate')

    fldChar3 = OxmlElement('w:fldChar')
    fldChar3.set(qn('w:fldCharType'), 'end')

    run._r.append(fldChar1)
    run._r.append(instrText)
    run._r.append(fldChar2)
    run._r.append(fldChar3)

    # Style key section titles to Headings so TOC can link them
    headings = ["PRELIMINARY STATEMENT", "JURISDICTION", "VENUE", "PARTIES", "FACTUAL ALLEGATIONS",
                "CAUSES OF ACTION", "DAMAGES", "PRAYER FOR RELIEF", "JURY DEMAND"]

    for para in doc.paragraphs:
        text = para.text.upper().strip()
        if text in headings:
            para.style = "Heading 1"

    doc.save(docx_path)
    print(f"‚úÖ TOC + headings applied to: {docx_path}")
```

```python
from docx import Document
from datetime import datetime
import os
import json # Import json here as it's used for loading and saving the case file

def generate_filing_cover_and_cert(case_name, plaintiff_name="[Your Name]"):
    with open(f"{case_name}.json") as f:
        case = json.load(f)

    meta = case.get("metadata", {})
    today = datetime.utcnow().strftime("%B %d, %Y")
    clerk = "Clerk of Court\nUnited States District Court\nSouthern District of Texas\nGalveston Division"
    title = "Filing Cover Letter & Certificate of Service"
    # Corrected the filename generation to properly count existing documents of this type
    existing = [d for d in case.get("documents", []) if d.get("type") == "filing_cover"]
    version = len(existing) + 1
    filename = f"filing_cover_and_cert_v{version}.docx"

    doc = Document()
    doc.add_paragraph(today)
    doc.add_paragraph("\n" + clerk)
    doc.add_paragraph(f"\nRE: {meta.get('case_style', 'Your Case Title')}")
    doc.add_paragraph("\nTo Whom It May Concern:\n")
    doc.add_paragraph("Please find enclosed the following documents submitted for filing in the above-referenced case:\n- [Insert documents here, e.g., Complaint, IFP Motion, Affidavit, etc.]\n")
    doc.add_paragraph("I respectfully request that the enclosed materials be docketed appropriately.\n\nSincerely,\n\n")
    doc.add_paragraph(f"{plaintiff_name}\nPro Se Plaintiff\n[Mailing Address Here]\n[Phone Number]\n[Email if applicable]")

    doc.add_paragraph("\n\n---\n\n")
    doc.add_paragraph("**Certificate of Service**")
    doc.add_paragraph(f"I certify that on {today}, I served a copy of the foregoing document on all parties or their counsel of record by [state method, e.g., certified mail, email, or in-person delivery].")

    doc.add_paragraph("\n_________________________\n" + plaintiff_name)

    doc.save(filename)

    # Save in case file
    # Ensure the 'documents' key exists before appending
    if "documents" not in case:
        case["documents"] = []
    case["documents"].append({
        "title": title,
        "filename": filename,
        "type": "filing_cover",
        "version": version,
        "timestamp": datetime.utcnow().isoformat()
    })
    with open(f"{case_name}.json", "w") as f:
        json.dump(case, f, indent=4) # Added indent for readability

    from google.colab import files
    files.download(filename)
    print(f"‚úÖ Filing letter and certificate saved: {filename}")
```
